```python
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
#to show matplotlib in jupyter notebook
%matplotlib inline
```

```python
df = pd.read_csv('heart.csv',parse_dates=True)
```

```python
df.head()
```

```python
df['Chol-Age'] = df.Chol - df.Age
df['Wynik'] = df['Chol-Age'].rolling(100).mean()
```

```python
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['Chol-Age'],df['Wynik'])
ax.set_xlabel('Index')
ax.set_ylabel('Chol-Age')
ax.set_zlabel('Wynik')
plt.show()
```

```python
#seaborn
plt.figure(figsize=(7,7), dpi = 100)
plt.title('SEABORN')
sns.boxplot( data=df,x='Age')
```
